# JOIN DALAM PYTHON


```python
# INNER JOIN (hanya mengambil yang beririsan di kedua tabel)
inner_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="inner")
print("Inner Join:\n",inner_join_df.head())

# LEFT JOIN (hanya mengambil semua data dari tabel kiri dan irisan dari kedua tabel)
left_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="left")
print("\nLeft Join:\n", left_join_df)

# RIGHT JOIN (hanya mengambil semua dari dari tabel kanan dan irisan dari kedua tabel)
right_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="right")
print("\nRight Join:\n", right_join_df)

# OUTER JOIN (mengambil semua kemungkinan dari kedua tabel, gabungkan jika cocok)
outer_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="outer")
print("\nOuter Join:\n", outer_join_df)
```

🔹 Penjelasan Metode Join

on="customer_id" → kolom yang digunakan untuk menggabungkan kedua DataFrame.

how parameter:

"inner" → hanya baris yang cocok di kedua DataFrame.

"left" → semua baris dari DataFrame kiri (customers_df) + data cocok dari kanan.

"right" → semua baris dari DataFrame kanan (orders_df) + data cocok dari kiri.

"outer" → semua baris dari kedua DataFrame, gabungkan jika ada kecocokan.
